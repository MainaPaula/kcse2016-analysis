```python
#importing the libraries required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-bright')
```

```python
# Importing the data
kcse_data = pd.read_csv('KCSE_2016.csv')
kcse_data
```

```python
#Sorting by gender
sorted_data = kcse_data.sort_values(by='Gender')
sorted_data.reset_index()
```

```python
#Barchart of 2016 gender performance
x_indexes = np.arange(len(sorted_data.iloc[2][1:].index))
width = 0.25

x_labels = sorted_data.iloc[2][1:].index

plt.xticks(x_indexes, x_labels)
plt.bar(x_indexes - width, sorted_data.iloc[20][1:], width = width, lw = 1.5, label='Male')
plt.bar(x_indexes, sorted_data.iloc[12][1:], width =width, lw = 1.5, label='Female')

plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.legend()
plt.title('Comparison of Performance in 2016 by gender')
plt.savefig('gender.png')
```

```python
#Barchart of 2015 versus 2016 performance
x_indexes = np.arange(len(sorted_data.iloc[2][1:].index))
width = 0.25

x_labels = sorted_data.iloc[2][1:].index

plt.xticks(x_indexes, x_labels)
plt.bar(x_indexes - width, sorted_data.iloc[5][1:], width = width, lw = 1.5, label='2015')
plt.bar(x_indexes, sorted_data.iloc[6][1:], width =width, lw = 1.5, label='2016')

plt.xlabel('Grades')
plt.ylabel('Number of Students')
plt.legend()
plt.title('Comparison of Performance between 2015 and 2016')
plt.savefig('2015-2016.png')
```

```python
new_data = sorted_data.drop('Gender',1)
new_data
```

```python
student_sum = new_data.sum(1)
student_sum
```

```python
#Piechart for gender demographics
sections = [student_sum.iloc[12],student_sum.iloc[20]]
labels = ['Female', 'Male']
colors = ['#008fd5', '#fc4f30']
plt.title('2016 Demographics by Gender')
plt.pie(sections, labels=labels, wedgeprops={'edgecolor':'black', }, colors = colors)
plt.savefig('gender_population2016.png')
```

```python
years = [2006,2007,2008,2009, 2010,2015,2016]

x_indexes = len(years)
width = 0.2

plt.xticks(years, years)

plt.bar(x_indexes - (3 * width), student_sum.iloc[0], width = width, lw = 1.5, label='2006')
plt.bar(x_indexes - (2 * width), student_sum.iloc[1], width = width, lw = 1.5, label='2007')
plt.bar(x_indexes - width, student_sum.iloc[2], width = width, lw = 1.5, label='2008')
plt.bar(x_indexes, student_sum.iloc[3], width = width, lw = 1.5, label='2009')
plt.bar(x_indexes + width, student_sum.iloc[4], width = width, lw = 1.5, label='2010')
plt.bar(x_indexes + (2*width), student_sum.iloc[5], width = width, lw = 1.5, label='2015')
plt.bar(x_indexes + (3*width), student_sum.iloc[6], width = width, lw = 1.5, label='2016')

plt.xlabel('Years')
plt.ylabel('Number of Students')
plt.title('Comparison of Performance by Total')
plt.legend()
plt.tight_layout()
plt.savefig('total_comparison.png')
```
